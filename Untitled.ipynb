```python
import pandas as pd
import seaborn as sns
sns.set_theme()
```

```python
lotus = pd.read_csv("./data/230106_frozen_metadata.csv.gz", low_memory=False)
lotus = lotus.rename(columns={'organism_name': 'species', 'structure_smiles_2D': 'molecule'})
```

```python
par_1 = pd.read_parquet("./out/chunk_0_result.parquet")
```

```python
par_2 = pd.read_parquet("./out/chunk_29_result.parquet")
```

```python
df = pd.concat([par_1, par_2])
```

```python
df_drop = df.merge(lotus, on=['species', 'molecule'], how='left', indicator=True)
```

```python
df_drop = df_drop[df_drop['_merge'] == 'left_only'].drop(columns='_merge')
```

```python
df_drop
```

```python
sns.histplot(data=df_drop, x=df_drop.prob, hue='species')
```

```python
df_drop[df_drop.prob>0.95]
```

```python
pc = pd.DataFrame([par_1.prob.to_list(),
                   par_2.prob.to_list()],
                  columns=[par_1.molecule.to_list()],
                 index=['Kalanchoe daigremontiana', 'Amaranthus hybridus'])
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
```

```python
pca_out = pca.fit_transform(pc)
```

```python
sns.scatterplot(x=pca_out[:,0], y=pca_out[:,1])
```
